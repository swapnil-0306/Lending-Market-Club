# Training & Testing Data

```python
from sklearn.model_selection import train_test_split
```

# Encode Categorical Columns

```python
import category_encoders as ce
```

```python
encoder = ce.OrdinalEncoder(cols=['term','int_rate','home_ownership','verification_status','issue_d','purpose','title','last_pymnt_d'])
```

```python
X_train = encoder.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = encoder.fit_transform(X_test)
```

```python
X_test
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(random_state=0)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Random Forest Model Accuracy score is ',accuracy_score(y_test,y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
gini_model = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
```

```python
gini_model.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10,8))
plot_tree(gini_model,feature_names=X_train.columns,filled=True)
plt.show()
```

```python
gini_model_pred=gini_model.predict(X_test)
```

```python
gini_model_pred
```

```python
print('Decision Tree Model Accuracy score is ',accuracy_score(y_test,gini_model_pred))
```

# SVM

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='sigmoid',C=1)
```

```python
clf.fit(X_train, y_train)
```

```python
svc_pred=clf.predict(X_test)
```

```python
svc_pred
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test, svc_pred))
```

```python
print(' SVM Model Accuracy score is ',accuracy_score(y_test,svc_pred))
```
